```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file, specifying the columns to be read
data = pd.read_excel('oee_ai_prep.xlsx', usecols=['Rejected Units', 'Schicht'])

# Group the data by weekday and sum the rejected units
rejected_by_weekday = data.groupby('Schicht')['Rejected Units'].sum()

# Create a bar chart
rejected_by_weekday.plot(kind='bar')

# Set chart labels and title
plt.xlabel('Schicht')
plt.ylabel('Rejected Units')
plt.title('Sum Rejected Units by Weekday')

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Pfad zur XLSX-Datei
xlsx_file = 'oee_ai_prep.xlsx'

# Lade die Daten aus der XLSX-Datei
df = pd.read_excel(xlsx_file)

# Füge eine neue Spalte hinzu, um den Wochentag zu extrahieren
df['Weekday'] = pd.to_datetime(df['Date']).dt.day_name()

# Berechne die Anzahl der Touch-Events pro Wochentag
touch_events_by_weekday = df.groupby('Weekday').apply(lambda x: x.filter(like='touch-events').sum().sum())

# Erstelle den Boxplot
touch_events_by_weekday.plot(kind='bar')

# Beschriftungen für den Boxplot
plt.title('Total Touch Events by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Total Touch Events')

# Zeige den Boxplot an
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Pfad zur XLSX-Datei
xlsx_file = 'oee_ai_prep.xlsx'

# Lade die Daten aus der XLSX-Datei
df = pd.read_excel(xlsx_file)

# Filtere die Spalten, die den Begriff "touch-events" enthalten
touch_event_columns = [col for col in df.columns if 'touch-events' in col.lower()]

# Berechne die Summe der Touch-Events pro Schicht
touch_events_by_shift = df.groupby('Schicht')[touch_event_columns].sum().sum()

# Erstelle den Boxplot
touch_events_by_shift.plot(kind='box')

# Beschriftungen für den Boxplot
plt.title('Total Touch Events by Shift')
plt.xlabel('Shift')
plt.ylabel('Total Touch Events')

# Zeige den Boxplot an
plt.show()
```
